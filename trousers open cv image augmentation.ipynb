```python
import cv2
import glob
import os
import random
import numpy as np
```

```python
#input folder is where our traning trousers images are stored(there are 80 photos in my training dataset will augment each photo)
inputfolder='C:/Users/COMP/men wear bottoms/train/trousers'
os.mkdir('flipped_t')#making a directory where are photos after the flip will be stored
i=0
for img in glob.glob(inputfolder+"/*jpeg"):
    image=cv2.imread(img)
    imgflipped=cv2.flip(image,1)
    cv2.imwrite("flipped_t/flip%04i.jpeg" %i,imgflipped)#saving my images in flipped folder and giving the photos label
    i +=1
    cv2.imshow('image',imgflipped)
    cv2.waitKey(30)
cv2.destroyAllWindows()
```

```python
#for rotation we have range (-30,30)so once I  will flip it 15(angle) to right and then 25(angle) to left and labelling the pictures and then saving them.
```

```python
inputfolder='C:/Users/COMP/men wear bottoms/train/trousers'
os.mkdir('rotation1_t')
i=0
for img in glob.glob(inputfolder+"/*jpeg"):
    image=cv2.imread(img)
    height,width=image.shape[:2]
    rotation_matrix=cv2.getRotationMatrix2D((width/2,height/2),-15,0.85)
    rotated_image=cv2.warpAffine(image,rotation_matrix,(width,height))
    cv2.imwrite("rotation1_t/rotatedr%04i.jpeg" %i,rotated_image)
    i +=1
    cv2.imshow('image',rotated_image)
    cv2.waitKey(30)
cv2.destroyAllWindows()
```

```python
inputfolder='C:/Users/COMP/men wear bottoms/train/trousers'
os.mkdir('rotation2_t')
i=0
for img in glob.glob(inputfolder+"/*jpeg"):
    image=cv2.imread(img)
    height,width=image.shape[:2]
    rotation_matrix=cv2.getRotationMatrix2D((width/2,height/2),25,0.85)
    rotated_image=cv2.warpAffine(image,rotation_matrix,(width,height))
    cv2.imwrite("rotation2_t/rotatedl%04i.jpeg" %i,rotated_image)
    i +=1
    cv2.imshow('image',rotated_image)
    cv2.waitKey(30)
cv2.destroyAllWindows()
```

```python
#similarly making folders for brightness_add,brightness_subtract and sharpening,reading the images applying the augmentaton and saving them
```

```python
inputfolder='C:/Users/COMP/men wear bottoms/train/trousers'
os.mkdir('brightness_add_t')
i=0
for img in glob.glob(inputfolder+"/*jpeg"):
    image=cv2.imread(img)
    M=np.ones(image.shape,dtype="uint8")*75
    added=cv2.add(image,M)
    cv2.imwrite("brightness_add_t/brightnessadd%04i.jpeg" %i,added)
    i +=1
    cv2.imshow('image',added)
    cv2.waitKey(30)
cv2.destroyAllWindows()
```

```python
inputfolder='C:/Users/COMP/men wear bottoms/train/trousers'
os.mkdir('brightness_subtract_t')
i=0
for img in glob.glob(inputfolder+"/*jpeg"):
    image=cv2.imread(img)
    M=np.ones(image.shape,dtype="uint8")*75
    subtracted=cv2.subtract(image,M)
    cv2.imwrite("brightness_subtract_t/brightnesssubtract%04i.jpeg" %i,subtracted)
    i +=1
    cv2.imshow('image',subtracted)
    cv2.waitKey(30)
cv2.destroyAllWindows()
```

```python
inputfolder='C:/Users/COMP/men wear bottoms/train/trousers'
os.mkdir('sharpening_t')
i=0
for img in glob.glob(inputfolder+"/*jpeg"):
    image=cv2.imread(img)
    kernel_sharpening=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
    sharpened=cv2.filter2D(image,-1,kernel_sharpening)
   
    cv2.imwrite("sharpening_t/sharpened%04i.jpeg" %i,sharpened)
    i +=1
    cv2.imshow('image',sharpened)
    cv2.waitKey(30)
cv2.destroyAllWindows()
```
